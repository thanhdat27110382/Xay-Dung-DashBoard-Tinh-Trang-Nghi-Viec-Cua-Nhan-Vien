```python
#Biểu đồ violin plot so sánh thu nhập hàng tháng của nhân viên theo giới tính, và phân loại thêm theo trạng thái nghỉ việc
# Chỉnh kích thước biểu đồ
plt.figure(figsize=(10, 6))  
ax = sns.violinplot(x ='Gender', y ='Monthly Income (USD)', data = data, hue ='Attrition (Yes/No)', split = True)
# Thêm giá trị trung bình lên biểu đồ
for i in range(len(income_stats)):
    gender = income_stats.loc[i, 'Gender']
    attrition = income_stats.loc[i, 'Attrition (Yes/No)']
    income_mean = income_stats.loc[i, 'Monthly Income (USD)']  # Giá trị trung bình của Monthly Income
    
    # Tìm vị trí trên biểu đồ để hiển thị số liệu
    x_pos = list(data['Gender'].unique()).index(gender)
    hue_pos = list(data['Attrition (Yes/No)'].unique()).index(attrition)
    
    # Thêm giá trị trung bình lên biểu đồ
    ax.text(x_pos + hue_pos * 0.2, income_mean, f'Mean: {income_mean:.2f}', ha='center', va='center', color='black', fontweight='bold')

# Thêm tiêu đề và nhãn cho biểu đồ
plt.title('Monthly Income (USD) theo Gender và Attrition (Yes/No)')
plt.xlabel('Gender')
plt.ylabel('Monthly Income (USD)')

# Hiển thị biểu đồ
plt.show()
```

### SMOTE Sampling

SMOTE (Synthetic Minority Oversampling Technique) là một kỹ thuật phổ biến trong phân tích dữ liệu để xử lý tình trạng mất cân bằng dữ liệu. Khi làm việc với các tập dữ liệu mà số lượng mẫu của lớp thiểu số (minority class) ít hơn đáng kể so với lớp đa số (majority class), SMOTE giúp tăng cường dữ liệu lớp thiểu số bằng cách tạo ra các mẫu tổng hợp mới.

```python
# Tách dữ liệu thành các phần:
# y2 là nhãn (label) mà mô hình cần dự đoán
# x2 sẽ là dữ liệu đầu vào cho mô hình học máy
from collections import Counter
y2 = data['Attrition (Yes/No)']
x2 = data.drop(['Attrition (Yes/No)'], axis = 1)
```

```python
# in biến y2
y2
```

```python
# in biến x2
x2
```

```python
from imblearn.combine import SMOTETomek
# sử dụng SMOTETomek để xử lý vấn đề mất cân bằng dữ liệu
smk = SMOTETomek()
x_smote, y_smote = smk.fit_resample(x2, y2)
```

```python
print('hình dạng tập dữ liệu gốc:', Counter(y2))
```

```python
print('Hình dạng tập dữ liệu lấy mẫu của SMOTE:', Counter(y_smote))
```

Ở đây, chúng ta có thể thấy số lượng "Yes" và "No" bằng nhau

### KNN

KNN (K-Nearest Neighbors) là một thuật toán học máy đơn giản và phổ biến, thuộc nhóm học không giám sát hoặc học có giám sát, tùy thuộc vào cách sử dụng. KNN được sử dụng chủ yếu trong các bài toán phân loại và hồi quy.

```python
#thực hiện quá trình huấn luyện và đánh giá mô hình KNN với các giá trị khác nhau của tham số k
knn_scores = []
for k in range(1,4):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    knn_classifier.fit(X_train,y_train)
    knn_scores.append(knn_classifier.score(X_test, y_test))
```

```python
#biểu đồ đường trực quan hóa mối quan hệ giữa số lượng Neighbors (K) và độ chính xác (scores) của mô hình KNN
plt.plot([k for k in range(1, 4)], knn_scores, color = 'red')
for i in range(1,4):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 4)])
plt.rcParams["figure.figsize"] = (15,15)
plt.xlabel('Số lượng Neighbors (K)')
plt.ylabel('Scores')
plt.title('Điểm số của Bộ phân loại K Neighbors cho các giá trị K khác nhau')
```

```python
# thực hiện dự đoán nhãn (labels) trên tập kiểm tra X_test bằng mô hình KNN đã huấn luyện, sau đó in báo cáo phân loại đánh giá hiệu suất của mô hình.
y_pred_knn = knn_classifier.predict(X_test)
print(classification_report(y_test, y_pred_knn))
```

```python
# ma trận confusion matrix trực quan hóa hiệu suất của mô hình KNN trên tập kiểm tra (y_test và y_pred_knn).
from sklearn import metrics
cm = confusion_matrix(y_test, y_pred_knn)
display = metrics.ConfusionMatrixDisplay(cm)
display.plot()
plt.show()
```

### Support Vector Machine

SVM (Support Vector Machine) là một thuật toán học máy mạnh mẽ được sử dụng chủ yếu cho các bài toán phân loại và hồi quy, đặc biệt là với các bài toán có không gian dữ liệu có số chiều cao.

SVC (Support Vector Classification) là một phiên bản của SVM (Support Vector Machine), được sử dụng chủ yếu cho các bài toán phân loại. SVC là thuật toán phân loại trong SVM, tìm kiếm hyperplane tối ưu để phân chia dữ liệu thành các lớp khác nhau trong không gian đặc trưng.

```python
# thực hiện huấn luyện và đánh giá một mô hình Support Vector Classifier (SVC)
classifier_svc = SVC()
classifier_svc.fit(X_train, y_train)
score = classifier_svc.score(X_test, y_test)
# in độ chính xác
print(score)
```

```python
#sử dụng mô hình SVC đã được huấn luyện để dự đoán nhãn trên tập kiểm tra (X_test).
y_pred_svc = classifier_svc.predict(X_test)
```

```python
# in 
y_pred_svc
```

```python
#tạo một báo cáo phân loại để đánh giá hiệu suất của mô hình SVC dựa trên nhãn thực tế (y_test) và nhãn dự đoán (y_pred_svc).
print(classification_report(y_test,y_pred_svc))
```

```python
#ma trận confusion matrix để đánh giá hiệu suất của mô hình SVC dựa trên nhãn thực tế (y_test) và nhãn dự đoán (y_pred_svc).
from sklearn import metrics
cm = confusion_matrix(y_test,y_pred_svc)
display = metrics.ConfusionMatrixDisplay(cm)
display.plot()
plt.show()
```

### Random forest

Random Forest là một thuật toán học máy mạnh mẽ thuộc nhóm học giám sát (supervised learning), được sử dụng cho các bài toán phân loại và hồi quy. Nó là một phương pháp ensemble learning, có nghĩa là nó kết hợp nhiều mô hình đơn lẻ để đưa ra kết quả cuối cùng. Trong trường hợp của Random Forest, các mô hình đơn lẻ là các Cây quyết định (Decision Trees).

```python
#sử dụng thuật toán Random Forest Classifier để xây dựng mô hình phân loại, huấn luyện mô hình trên tập dữ liệu huấn luyện
#và sau đó dự đoán nhãn trên tập dữ liệu kiểm tra.
random_forest = RandomForestClassifier()
random_forest.fit(X_train,y_train)
y_pred_RF = random_forest.predict(X_test)
```

```python
#tính toán độ chính xác (accuracy) của mô hình Random Forest trên tập kiểm tra, sau đó in kết quả dưới dạng phần trăm.
accuracy_RF = accuracy_score(y_test, y_pred_RF)
print("Độ chính xác của Random Forest", accuracy_RF * 100.0)
```

```python
#tạo báo cáo chi tiết về các chỉ số đánh giá hiệu suất của mô hình Random Forest dựa trên nhãn thực tế (y_test) và nhãn dự đoán (y_pred_RF).
print(classification_report(y_test, y_pred_RF))
```

```python
#ma trận confusion matrix đánh giá hiệu suất của mô hình Random Forest và hiển thị ma trận dưới dạng đồ thị.
cm = confusion_matrix(y_test, y_pred_RF)
display = metrics.ConfusionMatrixDisplay(cm)
display.plot()
plt.show()
```
